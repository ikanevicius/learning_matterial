```python
import random

class DebitCard:
    def __init__(
        self,
        card_number,
        pin_code,
        holder_name,
        holder_surname,
        balance,
        takeout_limit,
        transactions):

        self.card_number = card_number
        self.pin_code = int(pin_code)
        self.holder_name = holder_name
        self.holder_surname = holder_surname
        self.balance = float(balance)
        self.takeout_limit = int(takeout_limit)
        self.transactions = transactions


    def check_pin(self):
        """Asks to enter PIN code and checks if the entered PIN code is correct."""
        attempts_given = 3
        while attempts_given:
            pin_entered = int(input("Enter your PIN code: "))
            if pin_entered == self.pin_code:
                return
            elif pin_entered != self.pin_code:
                attempts_given -= 1
                print(f"PIN code entered wrong. Attempts left: {attempts_given}")
        else:
            print("You failed 3 times and can`t continue.")


    def update_transactions(self):
        """Checks transactions history and makes sure not more than 5 would be kept."""
        while len(self.transactions) > 3:
            transactions = self.transactions
            transactions.pop(0)
        else:
            pass


    def add_money(self):
        self.check_pin()
        money_added = float(input("How much money you want to add: "))
        balance = self.balance
        self.balance = balance + money_added
        transactions = list(self.transactions)
        money_added_info = "+" + str(money_added) + "€"
        transactions.append(money_added_info)
        self.transactions = transactions
        self.update_transactions()

        return self.balance, self.transactions


    def take_money(self):
        self.check_pin()
        money_taken = float(input("How much money you want to withdraw: "))
        balance = self.balance
        if money_taken <= balance:
            takeout_limit = self.takeout_limit
            if money_taken <= takeout_limit:
                self.balance = balance - money_taken
                transactions = list(self.transactions)
                money_taken_info = "-" + str(money_taken) + "€"
                transactions.append(money_taken_info)
                self.transactions = transactions
                self.update_transactions()
            elif money_taken > takeout_limit:
                print(f"Your can`t takeout more than {takeout_limit}€ at once.")
        elif money_taken > balance:
            print(f"""You don`t have enought money to withdraw {money_taken}€ from your account.""")

        return self.balance, self.transactions


    def change_takeout_limit(self):
        self.check_pin()
        set_new_limit = float(input("Enter your new takeout limit: "))
        self.takeout_limit = set_new_limit

        print(f"""Your takeout limit was changed succesfully! 
You can now takeout up to {self.takeout_limit}€ at once.
""")
        return self.takeout_limit


    def change_pin(self):
        attempts_given = 3
        while attempts_given:
            old_pin = int(input("Enter your old PIN code: "))
            if old_pin == self.pin_code:
                new_pin = int(input("Enter your new PIN code: "))
                if len(str(new_pin)) == 4 and new_pin != old_pin:
                    self.pin_code = new_pin
                    print(f"""Your PIN code was succesfully changed!
Your new PIN is {self.pin_code}.
""")
                    break
                elif len(str(new_pin)) == 4 and new_pin == old_pin:
                    print("New PIN code can`t bet the same as the old one.")
                    attempts_given -= 1
                elif len(str(new_pin)) != 4:
                    print("PIN code must contain 4 digits.")
                    attempts_given -= 1
            elif old_pin != self.pin_code:
                print(f"Old PIN code was entered wrong. Please try again.")
                attempts_given -= 1
        else:
            print("You failed 3 times and can`t change the PIN code now.")


def create_new_card():
    card_number = int(random.uniform(1000000,9999999))
    pin_code = int(random.uniform(1000,9999))
    holder_name = input("Enter card holder name: ")
    holder_surname = input("Enter card holder surname: ")
    balance = 0
    takeout_limit = 250
    transactions = []

    card_created = DebitCard(
        card_number,
        pin_code,
        holder_name,
        holder_surname,
        balance,
        takeout_limit,
        transactions
    )

    return card_created

card_1 = create_new_card()
print(f"""Dear {card_1.holder_name} {card_1.holder_surname}, your debit card was created succesfully!
Your card number is {card_1.card_number}, PIN: {card_1.pin_code}.
You will be able to take no more than {card_1.takeout_limit}€ at once.
""") # Nukelta į kitą eilutę, kad vartotojo sąsajoje atsirastų tarpas, taip palengvinant skaitymą.

card_1.change_takeout_limit()

card_1.add_money()
card_1.add_money()
card_1.take_money()
print(f"""Balance: {round(card_1.balance, 2)}€, 
Last 3 transactions made: {card_1.transactions}
""")

card_1.add_money()
print(f"""Balance: {round(card_1.balance, 2)}€, 
Last 3 transactions made: {card_1.transactions}
""")

card_1.take_money()
print(f"""Balance: {round(card_1.balance, 2)}€, 
Last 3 transactions made: {card_1.transactions}
""")

card_1.change_pin()

card_1.add_money()
print(f"""Balance: {round(card_1.balance, 2)}€, 
Last 3 transactions made: {card_1.transactions}
""")
```
